### Explore Parking lot in toronto

```python
#'Venue Address', 'Venue Distance',
def get_nearby_venues(names, latitudes, longitudes, radius = 500):
    venues_lst = []
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
        url_foursquare = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET,
        VERSION,
        lat,
        lng,
        radius,
        limit)

        response = requests.get(url_foursquare).json()['response']['groups'][0]['items']
    
    #return only relevant information for each nearby venue
    
        venues_lst.append([(
            name, lat, lng, item['venue']['name'],
            #item['venue']['location']['address'],
            item['venue']['location']['lat'],
            item['venue']['location']['lng'],       
            item['venue']['location']['distance'],
            item['venue']['categories'][0]['name']
        ) for item in response])

        v = pd.DataFrame([item for venues_lst in venues_lst for item in venues_lst])

        v.columns = ['Park Name', 'Parking Latitude', 'Parking Longitude', 'Venue','Venue Latitude', 'Venue Longitude', 'Distance' ,'Venue Category']
    
    return (v)
```

### Analyze each parking lot

```python
# one hot-encoding

toronto_ohe = pd.get_dummies(toronto_venues[['Venue Category']], prefix='', prefix_sep='')

#add parking lot to column back to dataframe
toronto_ohe['Park Name'] = parking_news['Park Name']

#moving Park Name column to the first column
fixed_columns = [toronto_ohe.columns[-1]] + list(toronto_ohe.columns[:-1])
# list(smt.columns[:-1]) => return all element except the last one
toronto_ohe = toronto_ohe[fixed_columns]
toronto_ohe.head(50)
```

### grouping rows by park name and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_ohe.groupby('Park Name').mean().reset_index()
toronto_grouped.head()
```

```python
#print each park along with the top 5 common venues
num_top_venues = 5

for park in toronto_grouped['Park Name']:
    print('----'+park+'----')
    # ....T => transpose df
    temp = toronto_grouped[toronto_grouped['Park Name'] ==  park].T.reset_index()
    temp.columns = ['venue','frequency']
    temp = temp.iloc[1:]
    temp['freq'] = temp['frequency'].astype(float)
    temp = temp.round({'frequency':2})
    print(temp.sort_values('frequency', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
    
```

```python
#sort the venues in descending order

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#creat the new dataframe and display the top 5 venues for each park

num_top_venues = 5
indicators = ['st','nd','rd']

#create columns according to number of top venues
columns = ['Park Name']
for i in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(i+1, indicators[i]))
    except:
        columns.append('{} Most common venue'.format(i+1))

#create new dataframe
data_sorted = pd.DataFrame(columns=columns)
data_sorted['Park Name'] = toronto_grouped['Park Name']

for i in np.arange(toronto_grouped.shape[0]):
    data_sorted.iloc[i,1:] = return_most_common_venues(toronto_grouped.iloc[i, :], num_top_venues)
    
data_sorted.head()
```

### clustering park

```python
from sklearn.cluster import KMeans

k = 5

toronto_grouped_clustering = toronto_grouped.drop('Park Name', 1)
kmeans = KMeans(n_clusters=k, random_state=0).fit(toronto_grouped_clustering)

#check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

### creating a new dataframe that includes the cluster as well as the top 10 venues of each park

```python
# add clustering labels
data_sorted.insert(0,'Cluster Labels', kmeans.labels_)

toronto_merged = parking_news.join(data_sorted.set_index('Park Name'), on='Park Name')
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

#set color scheme for the clusters

x = np.arange(k)
ys = [i + x +(i*x)**2 for x in range(k)]
colors_array = cm.rainbow(np.linspace(0,1,len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

#add markers to the map

markers_colors = []

for lat, lng, name, cluster in zip(toronto_merged['latitude'],toronto_merged['longitude']
                                 ,toronto_merged['Park Name'],toronto_merged['Cluster Labels']):
    label = folium.Popup(str(name) + 'Cluster' + str(cluster), parse_html=True)
    folium.CircleMarker([lat,lng],
                       radius = 5,
                       popup = label,
                       color =rainbow[cluster-1],
                       fill=True,
                       fill_color = rainbow[cluster-1],
                       fill_opacity = 0.7).add_to(map_clusters)
    
map_clusters
```
